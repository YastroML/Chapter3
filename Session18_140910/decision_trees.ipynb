```python
import matplotlib
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
```

# Decision Trees

First, let's load the data we will be playing with.  This is the SDSS spectroscopic galaxy sample.

```python
from astroML.datasets import fetch_sdss_specgals

# Fetch data and prepare it for the computation
data = fetch_sdss_specgals()

# put magnitudes in a N by 5 array
mag = np.vstack([data['modelMag_%s' % f] for f in 'ugriz']).T
z = data['z']
```

Now, I am going to plot the galaxy color vs redshift, just to make sure I understand the data.

```python
#just taking a look at the galaxies we are playing with...
plt.figure(figsize=(6,6))
plt.scatter(z,mag[:,1]-mag[:,2],s=1,c=mag[:,1]-mag[:,2],marker="o",lw=0,vmin=-0.5,vmax=2)
plt.xlabel(r'spec_z')
plt.ylabel('g-r')
plt.ylim([-0.5,2])
plt.xlim([0,0.6])
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor

# train on ~60,000 points
mag_train = mag[::10] #fancy indexing--take every tenth point
z_train = z[::10] #fancy indexing--take every tenth point

# test on ~6,000 separate points
mag_test = mag[1::100]#fancy indexing--take every 100th point
z_test = z[1::100]#fancy indexing--take every 100th point

#create tree on decision tree on training set
clf = DecisionTreeRegressor(max_depth=15, random_state=0)
clf.fit(mag_train, z_train)

z_fit = clf.predict(mag_test)

plt.figure(figsize=(6,6))
plt.scatter(z_test, z_fit, marker="o", s=2, lw=0, c='k')
plt.plot([-0.1, 0.4], [-0.1, 0.4], ':k')
plt.xlabel(r'$z_{\rm true}$')
plt.ylabel(r'$z_{\rm fit}$')
plt.xlim(-0.02, 0.4001)
plt.ylim(-0.02, 0.4001)
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor

clf = RandomForestRegressor(n_estimators=10,max_depth=20, random_state=0)
clf.fit(mag_train, z_train)

z_fit_train = clf.predict(mag_train)
z_fit = clf.predict(mag_test)
rms_train = np.mean(np.sqrt((z_fit_train - z_train) ** 2))
rms_test = np.mean(np.sqrt((z_fit - z_test) ** 2))

plt.figure(figsize=(6,6))
plt.scatter(z_test, z_fit, marker='o', s=2, lw=0, c='k')
plt.plot([-0.1, 0.4], [-0.1, 0.4], ':k')
plt.xlabel(r'$z_{\rm true}$')
plt.ylabel(r'$z_{\rm fit}$')
plt.xlim(-0.02, 0.4001)
plt.ylim(-0.02, 0.4001)
plt.show()
```
