# Chapter 9: Classification

We will try to quantify how different data live in different places in their domain. This is simmilar to our discussion of clustering, however now our data have labels. We know first hand that some of our data are intrisically different than others.

Once we define the boundaries of the different 'neighboorhoods' of our data, we can try to understand the relationship between the data and their attributes, and use boundaries to make predictions about which class a new data point belongs to.

We usually say that there are points which are in our class, and point that are not. All our points have a set of features, which we store in the vector X which can have many dimensions. The points in our class have a y=1 data label, and the ones which are not have a y=0 label. 

## Using Bayes' Rule to Seperate Classes

Recall Bayes' rule. We can use it to construct our classifier. We need to take our features and interpret them to see if a given point is in the class or not 

$$g(x) = \frac{p(\textbf{x}|y=1)p(y=1)}{p(\textbf{x}|y=1)p(y=1) + p(\textbf{x}|y=0)p(y=0)}  $$

with a change of notation

$$g(x) = \frac{\pi_{1}p_{1}(\textbf{x})}{\pi_{1}p_{1}(\textbf{x}) + \pi_{0}p_{0}(\textbf{x})} $$

notice that each of these terms is a product of two normalized probabilities. This means at the most, g(x) is one, and at the least it is 0. Exactly what we need for our classification scheme. However, it can take on all values in between. We will use the following scheme

if g(x) > 1/2, then y=1. otherwise, y=0

OR

if p(y = 1|x) > p(y=0|x) then y=1. otherwise, y=0

OR

if $$ \pi_{1}p_{1}(x) > \pi_{0}p_{0}(x) $$ then y=1. otherwise, y=0

Above, the x's represent vectors of all the features. Let us assume that the features are conditionally independent. That is to say if xi and xj represent the ith and jth feature in the vector of x

$$ p(x^{i},x^{j}|y_{k}) = p(x^{i}|y)p(x^{j}|y_{k}) $$

Applying this idea, we can seperate out all the features in their own p, and rewrite the above as

$$ = \frac{\prod_{i}p(x^{i}|y_{k})p(y_{k})}{\sum_{j}\prod_{i}p(x^{i}|y_{j})p(y_{j})} $$

now in order to decide how to class the data set we choose the yk which maximizes this expression. This is kind of like our discussion of maximum liklihood estimators

This creates a new problem. What are the p's and pis? Well, we can pretend everything is a gaussian. Furthermore, we assume that the features are conditionally independent. that's how you get a gaussian bayes classifier

$$ \hat{y} = \textrm{arg max}[ \textrm{Ln} \pi_{k} - (1/2) \sum_{i=1}^{N} 2\pi(\sigma_{k}^{i})^2 + \frac{(x^{i} - \mu_{k}^{i})^{2}}{(\sigma_{k}^{i})^{2}}] $$

#### Gaussian Naive Bayes Classifier in action: Warm up 

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.naive_bayes import GaussianNB

#make some fake data. Remember, we need to come up with the sigmas and the mus
np.random.seed(0)
mu1 = [1, 1]
cov1 = 0.3 * np.eye(2)

mu2 = [5, 3]
cov2 = np.eye(2) * np.array([0.4, 0.1])

X = np.concatenate([np.random.multivariate_normal(mu1, cov1, 100),
                    np.random.multivariate_normal(mu2, cov2, 100)])
y = np.zeros(200)
y[100:] = 1

#------------------------------------------------------------
# Fit the Naive Bayes classifier
clf = GaussianNB()
clf.fit(X, y)

# predict the classification probabilities on a grid
xlim = (-1, 8)
ylim = (-1, 5)
xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 71),
                     np.linspace(ylim[0], ylim[1], 81))
Z = clf.predict_proba(np.c_[xx.ravel(), yy.ravel()])
Z = Z[:, 1].reshape(xx.shape)

#------------------------------------------------------------
# Plot the results
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111)
ax.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.binary, zorder=2)

ax.contour(xx, yy, Z, [0.5], colors='k')

ax.set_xlim(xlim)
ax.set_ylim(ylim)

ax.set_xlabel('$x$')
ax.set_ylabel('$y$')

plt.show()
```

We need a way to quantitatively evaluate how well our model is doing at fitting the data. Often in classification, the benchmarks we usre are called 'completness' and 'contamination'

$$ \text{completness} = \frac{\text{true positives}}{\text{true positives} + \text{false negatives}} $$

$$ \text{contamination} = \frac{\text{false positives}}{\text{true positives} + \text{false positives}} $$

This will help us compare modesl to each other and weigh how effective they are

In the above example, we have completness=1 and contamination=0. In practice, it is very hard to get this *with out over fitting*, and data are not always so well seperated

Lets move onto an example with real data now. We will use a data set with photometric colors. They have been calssified already for which sources are RR lyrae and which are not. RR Lyrae tend to live in certain places on color-color diagrams, as we shall see

```python
from astroML.datasets import fetch_rrlyrae_combined
from astroML.utils import split_samples
from astroML.utils import completeness_contamination

#----------------------------------------------------------------------
# get data and split into training & testing sets
X, y = fetch_rrlyrae_combined()
X = X[:, [1, 0, 2, 3]]  # rearrange columns for better 1-color results
(X_train, X_test), (y_train, y_test) = split_samples(X, y, [0.75, 0.25],
                                                     random_state=0)

N_tot = len(y)
N_st = np.sum(y == 0)
N_rr = N_tot - N_st
N_train = len(y_train)
N_test = len(y_test)
N_plot = 5000 + N_rr

#----------------------------------------------------------------------
# perform Naive Bayes
classifiers = []
predictions = []
Ncolors = np.arange(1, X.shape[1] + 1)

order = np.array([1, 0, 2, 3])

for nc in Ncolors:
    clf = GaussianNB()
    clf.fit(X_train[:, :nc], y_train)
    y_pred = clf.predict(X_test[:, :nc])

    classifiers.append(clf)
    predictions.append(y_pred)

completeness, contamination = completeness_contamination(predictions, y_test)

print "completeness", completeness
print "contamination", contamination

#------------------------------------------------------------
# Compute the decision boundary
clf = classifiers[1]
xlim = (0.7, 1.35)
ylim = (-0.15, 0.4)

xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 81),
                     np.linspace(ylim[0], ylim[1], 71))

Z = clf.predict_proba(np.c_[yy.ravel(), xx.ravel()])
Z = Z[:, 1].reshape(xx.shape)

#----------------------------------------------------------------------
# plot the results
fig = plt.figure(figsize=(8,8))

# left plot: data and decision boundary
ax = fig.add_subplot(111)
im = ax.scatter(X[-N_plot:, 1], X[-N_plot:, 0], c=y[-N_plot:],
                s=4, lw=0, cmap=plt.cm.binary, zorder=2)
im.set_clim(-0.5, 1)

im = ax.imshow(Z, origin='lower', aspect='auto',
               cmap=plt.cm.binary, zorder=1,
               extent=xlim + ylim)
im.set_clim(0, 1.5)
ax.contour(xx, yy, Z, [0.5], colors='r')

ax.set_xlim(xlim)
ax.set_ylim(ylim)

ax.set_xlabel('$u-g$')
ax.set_ylabel('$g-r$')

plt.show()
```

### K Neighbors Classification

Another way we can go about this problem is to levereage the fact that our data have labels, and let that speak for itself. The places where data that are in our classification will define the boundary of their 'neighborhood'.

Imagine that we have a new point at some arbitrary spot on our color-color diagram, and we want to know how to classify it. We can look at the nearest neighbor, and assign our new point the same classification as its neighbor. 

Alternatively, we can use many more neighbors and let them vote among themselves to determine the classification of the new point. We can also weight the vote of a neighbor by its distance to the new point. This functions as a smoothing parameter, similar to our discussion of kernels last week.

bear in mind that you need the right kind of data to weight the votes by distance. It wont work if you have a more qualitative feature, like 'morphology' for example, for which we can not really compute a distance. 

In order for this method to work, we really need to have a well sampled parameter space with a lot of data. K neighbors can over fit data if there are 'sparse regions' in our data set. If there is a dearth of neighbors in a particular spot in our data set, K neighbors will not have enough information to evaluate the new point.

Now to draw the decision boundary, we use our algorithm over all* points the parameter space, and let the points vote on the classfication. the places in parameter space which are equally likely to be in the class and not in the class marks the decision boundary

```python
# Author: Jake VanderPlas
# License: BSD
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import colors

from sklearn.neighbors import KNeighborsClassifier
from astroML.datasets import fetch_rrlyrae_combined
from astroML.utils import split_samples
from astroML.utils import completeness_contamination

#----------------------------------------------------------------------
# get data and split into training & testing sets
X, y = fetch_rrlyrae_combined()
X = X[:, [1, 0, 2, 3]]  # rearrange columns for better 1-color results
(X_train, X_test), (y_train, y_test) = split_samples(X, y, [0.75, 0.25],
                                                     random_state=0)

N_tot = len(y)
N_st = np.sum(y == 0)
N_rr = N_tot - N_st
N_train = len(y_train)
N_test = len(y_test)
N_plot = 5000 + N_rr

#----------------------------------------------------------------------
# perform Classification

classifiers = []
predictions = []
Ncolors = np.arange(1, X.shape[1] + 1)
kvals = [1, 10]

for k in kvals:
    classifiers.append([])
    predictions.append([])
    for nc in Ncolors:
        clf = KNeighborsClassifier(n_neighbors=k)
        clf.fit(X_train[:, :nc], y_train)
        y_pred = clf.predict(X_test[:, :nc])

        classifiers[-1].append(clf)
        predictions[-1].append(y_pred)

completeness, contamination = completeness_contamination(predictions, y_test)

print "completeness", completeness
print "contamination", contamination

#------------------------------------------------------------
# Compute the decision boundary
clf = classifiers[1][1]
xlim = (0.7, 1.35)
ylim = (-0.15, 0.4)

xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 71),
                     np.linspace(ylim[0], ylim[1], 81))

Z = clf.predict(np.c_[yy.ravel(), xx.ravel()])
Z = Z.reshape(xx.shape)

#----------------------------------------------------------------------
# plot the results
fig = plt.figure(figsize=(8,8))

# left plot: data and decision boundary
ax = fig.add_subplot(111)
im = ax.scatter(X[-N_plot:, 1], X[-N_plot:, 0], c=y[-N_plot:],
                s=4, lw=0, cmap=plt.cm.binary, zorder=2)
im.set_clim(-0.5, 1)

im = ax.imshow(Z, origin='lower', aspect='auto',
               cmap=plt.cm.binary, zorder=1,
               extent=xlim + ylim)
im.set_clim(0, 2)

ax.contour(xx, yy, Z, [0.5], colors='r')

ax.set_xlim(xlim)
ax.set_ylim(ylim)

ax.set_xlabel('$u-g$')
ax.set_ylabel('$g-r$')


plt.show()
```

### Logistic Regression

As the name suggests, this is a regression approch to the problem. The function we use to predict the class of a point has the form

$$ p(y=1|x) = \frac{\text{exp}[\sum_{j}\theta{j}x^{j}]}{1 + \text{exp}[\sum_{j}\theta{j}x^{j}]} $$

here xj is the jth feature, and each feature has its own weight, theta. Now notice that as the argument of the exponential gets larger, this function quickly tends to 1. as the argument gets more negative, the function quickly tends to 0. When the argument is 0, the probability is 1/2. This suits our needs for classification fairly well.

the key now, is to choose the right thetas which minimize the classification error. 

The classification error was not explicitly defined in the text. However, one scheme that is sometimes used is shown below

If we denote our hypothesis of a training example as h(x), then

cost = -log(h(x)) when y=1
OR
cost = -log(1-h(x)) when y=0

notice that this cost function heavily peanalizes you when you are wrong. 

To pick the right theta's that minimize this cost function, you can use something like gradient decent or MCMC.

#### Logistic Regression Warm Up

```python
from sklearn.linear_model import LogisticRegression

#make some fake data. Remember, we need to come up with the sigmas and the mus
np.random.seed(0)
mu1 = [1, 1]
cov1 = 0.3 * np.eye(2)

mu2 = [5, 3]
cov2 = np.eye(2) * np.array([0.4, 0.1])

X = np.concatenate([np.random.multivariate_normal(mu1, cov1, 100),
                    np.random.multivariate_normal(mu2, cov2, 100)])
y = np.zeros(200)
y[100:] = 1

#------------------------------------------------------------
# Fit the logistic regression classifier
logr = LogisticRegression(penalty='l2')
logr.fit(X, y)

# predict the classification probabilities on a grid
xlim = (-1, 8)
ylim = (-1, 5)
xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 71),
                     np.linspace(ylim[0], ylim[1], 81))
Z = logr.predict_proba(np.c_[xx.ravel(), yy.ravel()])
Z = Z[:, 1].reshape(xx.shape)

#------------------------------------------------------------
# Plot the results
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111)
ax.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.binary, zorder=2)

ax.contour(xx, yy, Z, [0.5], colors='k')

ax.set_xlim(xlim)
ax.set_ylim(ylim)

ax.set_xlabel('$x$')
ax.set_ylabel('$y$')

plt.show()
```

As it is presented in the reading, we can only do a straight line to our data. If your data is well seperated, that is ok, and you wont have to worry about high contamination. However, in real life we do not usually deal with data that is well seperated, and linear models may not cut it. Instead we can do something called 'feature maping' where we make new features out of our old ones, simply by taking their squares, or high powers still. We can do the previous exercise over, and by only making changes to two lines we can have a more powerful result

```python
from sklearn.linear_model import LogisticRegression

#make some fake data. Remember, we need to come up with the sigmas and the mus
np.random.seed(0)
mu1 = [1, 1]
cov1 = 0.3 * np.eye(2)

mu2 = [5, 3]
cov2 = np.eye(2) * np.array([0.4, 0.1])

X = np.concatenate([np.random.multivariate_normal(mu1, cov1, 100),
                    np.random.multivariate_normal(mu2, cov2, 100)])
y = np.zeros(200)
y[100:] = 1

#------------------------------------------------------------
# Fit the logistic regression classifier
logr = LogisticRegression(penalty='l2')
logr.fit(np.column_stack((X,X*X)), y)          #CHANGE HERE!!!

# predict the classification probabilities on a grid
xlim = (-1, 8)
ylim = (-1, 5)
xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 71),
                     np.linspace(ylim[0], ylim[1], 81))
Z = logr.predict_proba(np.c_[xx.ravel(), yy.ravel(),(xx*xx).ravel(),(yy*yy).ravel()]) #CHANGE HERE TOO!!!
Z = Z[:, 1].reshape(xx.shape)

#------------------------------------------------------------
# Plot the results
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111)
ax.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.binary, zorder=2)

ax.contour(xx, yy, Z, [0.5], colors='k')

ax.set_xlim(xlim)
ax.set_ylim(ylim)

ax.set_xlabel('$x$')
ax.set_ylabel('$y$')

plt.show()
```
