```python
%matplotlib inline
from matplotlib.pyplot import *
from matplotlib import rcParams

rcParams['image.interpolation'] = 'none'
rcParams['figure.figsize'] = (10,10)
```

```python
import getdata
from sklearn import cluster

from astropy.utils import timer
from astropy.modeling import models
```

```python
getdata.download_data()
```

```python
cattab, readme = getdata.load_data()
print readme
```

```python
cattab.colnames
```

```python
ra = cattab['ra']
dec = cattab['dec']

z = cattab['z_peak']
haszspec = cattab['z_spec']!=-1
z[haszspec] = cattab['z_spec'][haszspec]
```

```python
fields = np.unique(cattab['field'])
print fields

fimsks = dict([(fi, cattab['field']==fi) for fi in fields])
```

```python
for i, fi in enumerate(fields):
    plt.subplot(3,2,i+1)
    plt.scatter(ra[fimsks[fi]], dec[fimsks[fi]], c=z[fimsks[fi]], lw=0,s=3,alpha=.18,vmin=-0.1,vmax=10)
    plt.colorbar()
    plt.title(fi)
plt.tight_layout()
```

```python
cmsk = fimsks['COSMOS']
plt.scatter(ra[cmsk], dec[cmsk], c=z[cmsk], lw=0,s=3,alpha=.18,vmin=-0.1,vmax=10)
```

# Affinity Propogation

```python
msk = cmsk&(z!=-99)
data = np.array([ra[msk], dec[msk], z[msk]]).T
```

```python
ap = cluster.AffinityPropagation()
```

```python
rtp = timer.RunTimePredictor(lambda n: ap.fit(data[:n]))
rtp.time_func([1,5,10,50,100,500,1000])
```

```python
rtp.do_fit(model=models.Polynomial1D(degree=2))
```

```python
print rtp.predict_time(len(data))
rtp.plot()
```

Looks like this will take ~forever

# Kmeans

```python
msk = cmsk&(z!=-99)
#data = np.array([ra[msk], dec[msk], z[msk]]).T
#rescale the data
def rescaled(d):
    return (d - np.min(d))/np.ptp(d)

data = np.array([rescaled(ra[msk]), rescaled(dec[msk]), rescaled(z[msk])]).T
```

```python
km = cluster.KMeans()
km.fit(data)
```

```python
plt.scatter(ra[msk], dec[msk], c=km.labels_,lw=0,s=3)
plt.xlabel('ra')
plt.ylabel('dec')
```

```python
plt.scatter(z[msk], ra[msk], c=km.labels_,lw=0,s=3)
plt.xlabel('z')
plt.ylabel('ra')
```

# DBSCAN

```python
msk = cmsk&(z!=-99)
#data = np.array([ra[msk], dec[msk], z[msk]]).T
#rescale the data
def rescaled(d):
    return (d - np.min(d))/np.ptp(d)

data = np.array([rescaled(ra[msk]), rescaled(dec[msk]), rescaled(z[msk])]).T
```

```python
db = cluster.DBSCAN(eps=.008)
```

```python
db.fit(data)
incluster = db.labels_!=-1
```

```python
plt.scatter(ra[msk][incluster], dec[msk][incluster], c=db.labels_[incluster], lw=0, s=3)
plt.colorbar()
plt.xlabel('ra')
plt.ylabel('dec')
```

```python
plt.scatter(z[msk][incluster], ra[msk][incluster], c=db.labels_[incluster],lw=0,s=3)
plt.colorbar()
plt.xlabel('z')
plt.ylabel('ra')
```
