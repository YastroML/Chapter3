```python
%matplotlib inline
import numpy as np 
import matplotlib.pyplot as plt 
from astropy.io import ascii
```

```python
lightcurve = ascii.read("LightCurve.txt")

#a quick way to visualize the data w matplotlib
plt.plot(lightcurve['jd'],lightcurve['mag'],'ro')
plt.gca().invert_yaxis #magnitudes are backwards
plt.show()
```

```python
def gaussian_basis(x, mu, sigma):
    return np.exp(-0.5 * ((x - mu) / sigma) ** 2)
```

```python
x=lightcurve['jd']
```

```python
timestep=np.linspace(lightcurve['jd'][0],lightcurve['jd'][-1],100)
```

```python
X = gaussian_basis(x[:, np.newaxis], timestep, 10) #wrap all the gausians in a maxtrix
```

```python
from sklearn.linear_model import Lasso
model=Lasso(fit_intercept=True, alpha=0.01)
model.fit(X, lightcurve['mag'])
w = model.coef_
times=np.linspace(lightcurve['jd'][0],lightcurve['jd'][-1],1000)
fit = model.predict(gaussian_basis(times[:,np.newaxis],timestep,10))
```

```python
plt.plot(times,fit,'ro')
plt.plot(lightcurve['jd'],lightcurve['mag'],'ro')
```
