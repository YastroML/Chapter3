# Chapter 6: Correlation Functions

The probability of finding a pair of points in two volume elements, $dV_1$ and $dV_2$, seperated by a distance $r$ is given by:

\begin{equation}
dP_{12} = \rho^2dV_1dV_2(1+\xi(r)).
\end{equation}

For a completely random distribution, $dP_{12} = \rho^2dV_1dV_2$, so in the equation above, $\xi(r)$ is the excess(or decrement) probability of finding a pair of points with a seperation $r$.  It must take a value bounded between $-1$ and $\infty$.  It is called the two point correlation function(2PCF). 

To measure the 2PCF you need to be able to do two things:

1. count the number of pairs of data points as a function of the seperation,
2. and calculate the volume of the paramter space accesable to the data.

The first step is a simple couting of pairs denoted by $DD(r)$, data-data pairs.

The second step is usually accomplished by monte-carlo integration of the volume occupied by the data.  This is done by placing random points in the same volume as the data occupy.  You would like to use a large number of random points, so the volume estimate you get has a smaller error contribution to the estimate of the 2PCF than the discrete number of data points will contribute.     

You can then directly measure three things:

1. DD(r)  data-data
2. DR(r)  data-random
3. RR(r)  random-random

There are a few equations that are commonly used to estimate the 2PCF.  The standard method:

\begin{equation}
\xi(r)=\frac{DD(r)}{RR(r)}-1,
\end{equation}

and the very popular Landy-Szalay estimator:

\begin{equation}
\xi(r)=\frac{DD(r)-2DR(r)+RR(r)}{RR(r)}.
\end{equation}

Notice that the second is equivalent to the first if $DR = RR$, which it should, because there should be no correlation between the data and the randoms.

Lets take a look at my stupid brute force "mostly non-pythony" calculation of the 2PCF

```python
import sys

def auto_pair_counts_3D(data):
    N = len(data)
    dd = np.zeros((N**2,)) #store radial pair seperations 
    for i in range(0,N): #calculate distance between every point and every other point
        print "\r",i,
        sys.stdout.flush()
        x1 = data[i,0]
        y1 = data[i,1]
        z1 = data[i,2]
        x2 = data[:,0]
        y2 = data[:,1]
        z2 = data[:,2]
        dx = np.fabs(x1-x2)
        dy = np.fabs(y1-y2)
        dz = np.fabs(z1-z2)
        dd[i*N:i*N+N] = np.sqrt(dx**2.0+dy**2.0+dz**2.0) #distance formula
    return dd

def auto_pair_counts_2D(data):
    N = len(data)
    dd = np.zeros((N**2,)) #store radial pair seperations 
    for i in range(0,N): #calculate distance between every point and every other point
        print "\r",i,
        sys.stdout.flush()
        x1 = data[i,0]
        y1 = data[i,1]
        x2 = data[:,0]
        y2 = data[:,1]
        dx = np.fabs(x1-x2)
        dy = np.fabs(y1-y2)
        dd[i*N:i*N+N] = np.sqrt(dx**2.0+dy**2.0) #distance formula
    return dd

def cross_pair_counts_2D(data1,data2):
    N1 = len(data1)
    N2 = len(data2)
    dd = np.zeros((N1*N2,)) #store radial pair seperations 
    for i in range(0,N1): #calculate distance between every point and every other point
        print "\r",i,
        sys.stdout.flush()
        x1 = data1[i,0]
        y1 = data1[i,1]
        x2 = data2[:,0]
        y2 = data2[:,1]
        dx = np.fabs(x1-x2)
        dy = np.fabs(y1-y2)
        dd[i*N2:i*N2+N2] = np.sqrt(dx**2.0+dy**2.0) #distance formula
    return dd

def DC_two_point_3D(data,data_R,bins,method='standard'):
    factor = len(data_R) * 1.0 / len(data)
    dd = auto_pair_counts_3D(data)
    dd = np.histogram(dd,bins=bins)[0] #number of pairs with seperation between r and r+dr
    rr = auto_pair_counts_3D(data_R)
    rr = np.histogram(rr,bins=bins)[0] #number of pairs with seperation between r and r+dr
    if method == 'landy-szalay':
        dr = cross_pair_counts_2D(data,data_R)
        dr = np.histogram(dr,bins=bins)[0] #number of pairs with seperation between r and r+dr
        corr = (factor**2.0 * dd - 2.0 * factor * dr + rr) / rr
    else:
        corr = factor**2.0 * dd/rr - 1.0
    return corr

def DC_two_point_2D(data,data_R,bins,method='standard'):
    factor = len(data_R) * 1.0 / len(data)
    dd = auto_pair_counts_2D(data)
    dd = np.histogram(dd,bins=bins)[0] #number of pairs with seperation between r and r+dr
    rr = auto_pair_counts_2D(data_R)
    rr = np.histogram(rr,bins=bins)[0] #number of pairs with seperation between r and r+dr
    if method == 'landy-szalay':
        dr = cross_pair_counts_2D(data,data_R)
        dr = np.histogram(dr,bins=bins)[0] #number of pairs with seperation between r and r+dr
        corr = (factor**2.0 * dd - 2.0 * factor * dr + rr) / rr
    else:
        corr = factor**2.0 * dd/rr - 1.0
    #corr = np.nan_to_num(corr)
    return corr
     
```

Let's take a look at the code used by the book to calculate the 2PCF.

```python
#This is code I have taken from AstroML and modified to make it a bit clearer!
#I also added the ability to pass a random sample to the bootstrap function

##########Ignore this stuff here....#####################################
import warnings
from sklearn.neighbors import BallTree
from astroML.utils import check_random_state
# Check if scikit-learn's two-point functionality is available.
# This was added in scikit-learn version 0.14
try:
    from sklearn.neighbors import KDTree
    sklearn_has_two_point = True
except ImportError:
    import warnings
    sklearn_has_two_point = False  
print 'sklearn has 2-point correlation function:', sklearn_has_two_point
##########################################################################

def two_point(data, bins, method='standard', data_R=None, random_state=None):
    """Two-point correlation function
    Parameters
    ----------
    data : array_like
        input data, shape = [n_samples, n_features]
    bins : array_like
        bins within which to compute the 2-point correlation.
        shape = Nbins + 1
    method : string
        "standard" or "landy-szalay".
    data_R : array_like (optional)
        if specified, use this as the random comparison sample
    random_state : integer, np.random.RandomState, or None
        specify the random state to use for generating background
    Returns
    -------
    corr : ndarray
        the estimate of the correlation function within each bin
        shape = Nbins
    """
    
    data = np.asarray(data)
    bins = np.asarray(bins)
    rng = check_random_state(random_state)

    #check to see if there was improper input
    if method not in ['standard', 'landy-szalay']:
        raise ValueError("method must be 'standard' or 'landy-szalay'")
    if bins.ndim != 1:
        raise ValueError("bins must be a 1D array")
    if data.ndim == 1:
        data = data[:, np.newaxis]
    elif data.ndim != 2:
        raise ValueError("data should be 1D or 2D")

    n_samples, n_features = data.shape
    Nbins = len(bins) - 1

    # shuffle all but one axis to get background distribution
    if data_R is None:
        data_R = data.copy()
        for i in range(n_features - 1):
            rng.shuffle(data_R[:, i])
    else:
        data_R = np.asarray(data_R)
        if (data_R.ndim != 2) or (data_R.shape[-1] != n_features):
            raise ValueError('data_R must have same n_features as data')

    factor = len(data_R) * 1. / len(data)

    # The two-point correlation functions added in scikit-learn v. 0.14 is faster
    if sklearn_has_two_point:
        #first place both the data and the randoms into KD tree structures
        KDT_D = KDTree(data)
        KDT_R = KDTree(data_R)
        #then calculate the 2PCF using the trees...
        counts_DD = KDT_D.two_point_correlation(data, bins)
        counts_RR = KDT_R.two_point_correlation(data_R, bins)
    
    #but if you dont have an up-to-date scikit-learn use the AstroML version.
    else:
        warnings.warn("Version 0.3 of astroML will require scikit-learn "
                      "version 0.14 or higher for correlation function "
                      "calculations. Upgrade to sklearn 0.14+ now for much "
                      "faster correlation function calculations.")
        #first place both the data and the randoms into Ball tree structures
        BT_D = BallTree(data)
        BT_R = BallTree(data_R)
        #then create arrays to store the results
        counts_DD = np.zeros(Nbins + 1)
        counts_RR = np.zeros(Nbins + 1)
        #then for each bin, count the number of pairs that contribute using the tree
        for i in range(Nbins + 1):
            counts_DD[i] = np.sum(BT_D.query_radius(data, bins[i], count_only=True))
            counts_RR[i] = np.sum(BT_R.query_radius(data_R, bins[i], count_only=True))
    #becuase we counted all pints with seperations r<R, we need to take the differences for each bin
    DD = np.diff(counts_DD) #DD[i+1]-DD[i]
    RR = np.diff(counts_RR) #RR[i+1]-RR[i]

    # check for zeros in the denominator
    RR_zero = (RR == 0) #mask array ==True where there is a zero
    RR[RR_zero] = 1 #wherever there is a zero, set it equal to one instead to avoid annoying error messages

    #now use the estimators to calculate the 2PCF
    if method == 'standard':
        corr = factor ** 2.0 * DD / RR - 1.0
    elif method == 'landy-szalay':
        if sklearn_has_two_point:
            counts_DR = KDT_R.two_point_correlation(data, bins)
        else:
            counts_DR = np.zeros(Nbins + 1)
            for i in range(Nbins + 1):
                counts_DR[i] = np.sum(BT_R.query_radius(data, bins[i], count_only=True))
        DR = np.diff(counts_DR)
        corr = (factor ** 2 * DD - 2 * factor * DR + RR) / RR

    corr[RR_zero] = np.nan #wherever RR==0, set the 2PCF to nan

    return corr


def bootstrap_two_point(data, bins, Nbootstrap=10, method='standard', return_bootstraps=False, random_state=None, data_R=None):
    """Bootstrapped two-point correlation function
    Parameters
    ----------
    data : array_like
        input data, shape = [n_samples, n_features]
    bins : array_like
        bins within which to compute the 2-point correlation.
        shape = Nbins + 1
    Nbootstrap : integer
        number of bootstrap resamples to perform (default = 10)
    method : string
        "standard" or "landy-szalay".
    return_bootstraps: bool
        if True, return full bootstrapped samples
    random_state : integer, np.random.RandomState, or None
        specify the random state to use for generating background
    Returns
    -------
    corr, corr_err : ndarrays
        the estimate of the correlation function and the bootstrap
        error within each bin. shape = Nbins
    """
    
    data = np.asarray(data)
    bins = np.asarray(bins)
    rng = check_random_state(random_state)

    if method not in ['standard', 'landy-szalay']:
        raise ValueError("method must be 'standard' or 'landy-szalay'")

    if bins.ndim != 1:
        raise ValueError("bins must be a 1D array")

    if data.ndim == 1:
        data = data[:, np.newaxis]
    elif data.ndim != 2:
        raise ValueError("data should be 1D or 2D")

    if Nbootstrap < 2:
        raise ValueError("Nbootstrap must be greater than 1")

    n_samples, n_features = data.shape

    # get the baseline estimate
    corr = two_point(data, bins, data_R=data_R, method=method, random_state=rng)

    bootstraps = np.zeros((Nbootstrap, len(corr)))

    for i in range(Nbootstrap):
        indices = rng.randint(0, n_samples, n_samples)
        bootstraps[i] = two_point(data[indices, :], bins, data_R=data_R, method=method, random_state=rng)

    # use masked std dev in case of NaNs
    corr_err = np.asarray(np.ma.masked_invalid(bootstraps).std(0, ddof=1))

    if return_bootstraps:
        return corr, corr_err, bootstraps
    else:
        return corr, corr_err
```

```python
def uniform_sphere(RAlim, DEClim, size=1):
    """Draw a uniform sample on a sphere
    Parameters
    ----------
    RAlim : tuple
        select Right Ascension between RAlim[0] and RAlim[1]
        units are degrees
    DEClim : tuple
        select Declination between DEClim[0] and DEClim[1]
    size : int (optional)
        the size of the random arrays to return (default = 1)
    Returns
    -------
    RA, DEC : ndarray
        the random sample on the sphere within the given limits.
        arrays have shape equal to size.
    """
    
    zlim = np.sin(np.pi * np.asarray(DEClim) / 180.)
    z = zlim[0] + (zlim[1] - zlim[0]) * np.random.random(size)
    DEC = (180. / np.pi) * np.arcsin(z)
    RA = RAlim[0] + (RAlim[1] - RAlim[0]) * np.random.random(size)

    return RA, DEC


def ra_dec_to_xyz(ra, dec):
    """Convert ra & dec to Euclidean points
    Parameters
    ----------
    ra, dec : ndarrays
    Returns
    x, y, z : ndarrays
    """
    
    sin_ra = np.sin(ra * np.pi / 180.)
    cos_ra = np.cos(ra * np.pi / 180.)

    sin_dec = np.sin(np.pi / 2 - dec * np.pi / 180.)
    cos_dec = np.cos(np.pi / 2 - dec * np.pi / 180.)

    return (cos_ra * sin_dec, sin_ra * sin_dec, cos_dec)


def angular_dist_to_euclidean_dist(D, r=1):
    """convert angular distances to euclidean distances"""
    return 2 * r * np.sin(0.5 * D * np.pi / 180.)

def two_point_angular(ra, dec, bins, method='standard', random_state=None):
    """Angular two-point correlation function
    A separate function is needed because angular distances are not
    euclidean, and random sampling needs to take into account the
    spherical volume element.
    Parameters
    ----------
    ra : array_like
        input right ascention, shape = (n_samples,)
    dec : array_like
        input declination
    bins : array_like
        bins within which to compute the 2-point correlation.
        shape = Nbins + 1
    method : string
        "standard" or "landy-szalay".
    random_state : integer, np.random.RandomState, or None
        specify the random state to use for generating background
    Returns
    -------
    corr : ndarray
        the estimate of the correlation function within each bin
        shape = Nbins
    """
    
    ra  = np.asarray(ra)
    dec = np.asarray(dec)
    rng = check_random_state(random_state)

    if method not in ['standard', 'landy-szalay']:
        raise ValueError("method must be 'standard' or 'landy-szalay'")
    if bins.ndim != 1:
        raise ValueError("bins must be a 1D array")
    if (ra.ndim != 1) or (dec.ndim != 1) or (ra.shape != dec.shape):
        raise ValueError('ra and dec must be 1-dimensional '
                         'arrays of the same length')

    n_features = len(ra)
    Nbins = len(bins) - 1

    # draw a random sample with N points
    ra_R, dec_R = uniform_sphere((min(ra), max(ra)), (min(dec), max(dec)), 2 * len(ra))

    data = np.asarray(ra_dec_to_xyz(ra, dec), order='F').T
    data_R = np.asarray(ra_dec_to_xyz(ra_R, dec_R), order='F').T

    # convert spherical bins to cartesian bins
    bins_transform = angular_dist_to_euclidean_dist(bins)

    return two_point(data, bins_transform, method=method, data_R=data_R, random_state=rng)


def bootstrap_two_point_angular(ra, dec, bins, method='standard', Nbootstraps=10, random_state=None):
    """Angular two-point correlation function
    A separate function is needed because angular distances are not
    euclidean, and random sampling needs to take into account the
    spherical volume element.
    Parameters
    ----------
    ra : array_like
        input right ascention, shape = (n_samples,)
    dec : array_like
        input declination
    bins : array_like
        bins within which to compute the 2-point correlation.
        shape = Nbins + 1
    method : string
        "standard" or "landy-szalay".
    Nbootstraps : int
        number of bootstrap resamples
    random_state : integer, np.random.RandomState, or None
        specify the random state to use for generating background
    Returns
    -------
    corr : ndarray
        the estimate of the correlation function within each bin
        shape = Nbins
    dcorr : ndarray
        error estimate on dcorr (sample standard deviation of
        bootstrap resamples)
    bootstraps : ndarray
        The full sample of bootstraps used to compute corr and dcorr
    """
    
    ra = np.asarray(ra)
    dec = np.asarray(dec)
    rng = check_random_state(random_state)

    if method not in ['standard', 'landy-szalay']:
        raise ValueError("method must be 'standard' or 'landy-szalay'")
    if bins.ndim != 1:
        raise ValueError("bins must be a 1D array")
    if (ra.ndim != 1) or (dec.ndim != 1) or (ra.shape != dec.shape):
        raise ValueError('ra and dec must be 1-dimensional '
                         'arrays of the same length')

    n_features = len(ra)
    Nbins = len(bins) - 1
    data = np.asarray(ra_dec_to_xyz(ra, dec), order='F').T

    # convert spherical bins to cartesian bins
    bins_transform = angular_dist_to_euclidean_dist(bins)

    bootstraps = []

    for i in range(Nbootstraps):
        # draw a random sample with N points
        ra_R, dec_R = uniform_sphere((min(ra), max(ra)),
                                     (min(dec), max(dec)),
                                     2 * len(ra))

        data_R = np.asarray(ra_dec_to_xyz(ra_R, dec_R), order='F').T

        if i > 0:
            # random sample of the data
            ind = np.random.randint(0, data.shape[0], data.shape[0])
            data_b = data[ind]
        else:
            data_b = data

        bootstraps.append(two_point(data_b, bins_transform, method=method,
                                    data_R=data_R, random_state=rng))

    bootstraps = np.asarray(bootstraps)
    corr = np.mean(bootstraps, 0)
    corr_err = np.std(bootstraps, 0, ddof=1)

    return corr, corr_err, bootstraps
```
