## 7.4 Nonnegative Matrix Factorization

When using PCA, the resulting eigenvectors are defined relative to the mean data vector, resulting in principal componets that can negative or positive. However, in many physical systems, we know that the data vector can be represented by a linaer sum of only positive components. Example: a galaxy spectrum is a linear sum of stellar components, all with nonnegative contributions.

NMF applies an additional constraints on the components that they must all be positive. 

```python
from sklearn.decomposition import NMF

X = np.random.random((100,3))
nmf = NMF(n_components=3)
nmf.fit(X)
proj = nmf.transform(X)

comp = nmf.components_ # 3x10 array of components
err = nmf.reconstruction_err_ # how well 3 components capture the data
```

## 7.5 Manifold Learning

Not all real world data sets are linear and, therefore, we need an alternate, non-linear technique to compress and characterize these data sets. Manifold learning comprises as set of techniques that aim to accomplish non-linear dimensionality reduction. It allows non-linear surfaces to be unwrapped or unfolded so that the underlying structure becomes clear. In the following we will discuss two of the most popular manifold learning techniques:

- Locally Linear Embedding
- IsoMap (isometric mapping)

#### 7.5.1 Locally Linear Embedding

The key to locally linear embedding is that it attempts the dimensionality reduction with preserving the geometry of local neighborhoods of each points (relationship of each point to its $k$ nearest neighbors. It does this in two steps.

Step 1.

If you have a N points, you can describe their geometry using an $N \times\ N$ matrix, $W$, where each row describes each point in terms of all of the other points. LLE takes this one step further and zeros out all $W_{ij}$ except for the $k$ nearest neighbors. 

Step 2.

You take the weight matrix from Step 1 and use it in when minimizing the reconstruction error

$\epsilon(Y) = |Y - WY|^2 $

to find a matrix $Y$ of dimension $N \times\ d$ where $d$ is the dimension of the embedded manifold.

The solution to this can be obtained analytically using linear algebra techniques. 

```python
from sklearn.manifold import LocallyLinearEmbedding

X = np.random.normal(size=(1000,2)) # 100 points in 2 dim
R = np.random.random((2,10)) # projection matrix
X = np.dot(X,R) # now a 2D linear manifold in 10D space

k = 5 #num of nearest neighbors
n = 2 # number of dim in the fit

lle = LocallyLinearEmbedding(k,n)
lle.fit(X)
proj = lle.transform(X) #100 x 2 projection of the data
```

### 7.5.2 IsoMap

IsoMap is based on a multidimensional scaling framework. IsoMap recovers nonlinear structure by approximating geodesic curves which lie within the embedded manifold.

Step 1.

Create a connected graph, $G$, where $G_{ij}$ is the distance between point $i$ and $j$ if they are neighbors (else 0).

Step 2.

Construct a distance matrix $D_x$ which contains the length of the shortest path between $i$ and $j$ traversing $G$. 

```python
from sklearn.manifold import Isomap

X = np.random.normal(size=(1000,2)) # 100 points in 2 dim
R = np.random.random((2,10)) # projection matrix
X = np.dot(X,R) # now a 2D linear manifold in 10D space

k = 5 #num of nearest neighbors
n = 2 # number of dim in the fit


isomap = Isomap(k,n)
iso.fit(X)
proj = iso.transform(X)
```

### 7.5.3 Weaknesses of Manifold Learning

- Noisy and gappy data is problematic
- Likewise, manifold learning is sensitive to outliers
- Tuning parameters: How to choose the optimal selection method for neighbors has not been well constrained yet.
- The choice of output dimensionality is a free parameter with no solid way of estimating it yet. d = 1, 2 or 3 are commonly used, but only for practical reasons (ease of visualization).
- Reconstruction from the manifold is not very good.

## 7.6 Independent Component Analysis and Projection Pursuit

Independent component analysis (ICA) is a computational technique designed to solve the "cocktail party problem" : imagine a room with multiple microphones placed around a room with N people talking. Each microphone picks up a linear combination of the N voices. ICA is designed to use the concept of statistical independence to unmix the individual signals.

Each dataset can described as 

$x_1(k) = a_{11}s_1(k) + a_{12}s_2(k) + a_{13}s_3(k) + ...,$
$x_2(k) = a_{21}s_1(k) + a_{22}s_2(k) + a_{23}s_3(k) + ...,$
$x_3(k) = a_{31}s_1(k) + a_{32}s_2(k) + a_{33}s_3(k) + ...,$

where $s_i$ is each individual input signal and $a_{ij}$ is the appropriate mixing ampilitude. The key behind ICA is (1) the input signals are all statistically independent and, in most implementations, (2) the sum of two independent random variables will always be more Gaussian than either of the individual random variables (from the central limit theorem). ICA therefore looks for an unmixing matrix, W, that maximizes the non-Gaussianity of the distributions to isolate the distinct input signals.

```python
from sklearn.decomposition import FastICA

X = np.random.normal(size=(100,2))
R = np.random.random((2,5))
X = np.dot(X,R)

ica = FastICA(2) # fit two components
proj = ica.fit_transform(X)

comp = ica.components_ #the 2x5 matrix of independent components
sources = ica.sources_ # the 100x2 matrix of sources
```

The difference between ICA and projection pursuits is that projection pursuits seeks the "interesting" directions one at a time instead of simultaneously.

In astronomy, an example where you would use ICA is to decompose a spectrum of a galaxy where the individual input singals are the different stellar types.
