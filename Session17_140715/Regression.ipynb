# Principal Component Regression

Regression based on the principal components for high dimensional data sets.

Reminder: in PC<B>A</B> we find a set of orthogonal basis vectors and order them by the degree to which they minimize the variance in the sample.  We then, generally, discard those vectors which contribute little to describing the data.

Ridge Regression (from last week) does a very similar thing by using a cost function which eliminates components whose squared coefficients exceed some value.

PC<b>R</b> is useful when there are large covariances between some of your parameters.

<ul><li>Center the data</li>
<li>Perform Principal Component Analysis</li>
<li>Do a <i>linear</i> fit to the data in the PCA coordinate system</li>
<li>Convert the fit back to the data coordinates</li>
</ul>

# Kernel Regression

```python
import numpy as np
from astroML.linear_model import NadarayaWatson

import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline
mpl.rcParams['figure.figsize']=(10.,8.)
mpl.rcParams['font.size']=16
```

### Using NadarayaWatson to fit data which depends on two parameters

```python
# Generate a training set of data
X = np.random.random((100, 2)) # generate 100 points in 2 dimensions
y = X[:, 0] + X[:, 1]          # y is a linear combination of the two X 'parameters'

model = NadarayaWatson('gaussian', 0.05) # Instantiate a gaussian Kernel model with width 0.05
model.fit(X, y)                # use gaussian kernel regression to fit the training data

y_pred = model.predict(X)      # Given the same inputs, generate predicted y values using our model
```

```python
# Plot the training set y values
fig = plt.figure(1)
ax1 = fig.add_subplot(1,1,1)
ax1.set_xlabel("Point Number")
ax1.set_ylabel("y")
ax1.set_title("Model & Data")

# Original Data
pData = ax1.plot(y, color='#cccccc', marker='o',linestyle='',markeredgecolor='#cccccc')

# Overplot the predictions using our model
pModel = ax1.plot(y_pred, color='#cc0000', marker='+', linestyle='',markeredgewidth=2)
```

### Our old friend the cosmology data

```python
from astroML.cosmology import Cosmology
from astroML.datasets import generate_mu_z

# ------------- Generate the data to train on
z_sample, mu_sample, dmu = generate_mu_z(100, random_state=0)

# now the 'observed' data
cosmo = Cosmology()
z = np.linspace(0.01, 2, 1000)
mu_true = np.asarray(map(cosmo.mu, z))

# ------------- Initialize the classifier
kernelWidth = 0.1
muModel = NadarayaWatson('gaussian', h=kernelWidth)
nModelConstraints = 1 # just a weighted mean

muModel.fit(z_sample[:,None],mu_sample, dmu) # fit the sample data

mu_sample_fit = muModel.predict(z_sample[:,None]) # predict new mu from our sample z
mu_fit = muModel.predict(z[:,None])

chi2_dof_s = (np.sum(((mu_sample_fit - mu_sample) / dmu) ** 2) / (len(mu_sample) - nModelConstraints))
```

```python
print("Reduced Chi^2 = %.3f" % (chi2_dof_s))
```

```python
# ------------- Plot up the results
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
ax1.set_xlabel(r'$z$')
ax1.set_ylabel(r'$\mu$')

pFit = ax1.plot(z, mu_fit, linestyle='-', color='#000000')    # Predictions of mu based on z in black
pTrue = ax1.plot(z, mu_true, linestyle='--', color='#FF3333', linewidth=2)  # actual mu from cosmology data in red
# sample data with error bars
pData = ax1.errorbar(z_sample,mu_sample,dmu, marker='.', linestyle='', color='#3333CC', linewidth=1)
```

The kernel width (here 0.1) acts as a smoothing function so that in areas of low data density the fit doesn't get pulled too far off.  However, high density of points at ~z=0.25 weights those points more and pulls fit off.  You can vary h as a function of z to correct for this.

# Locally Linear Regression

Related to Kernel regression but instead of a locally constant solution at each point, a linear one.  Can increase the polynomial order, but going past linear apparently increases variance without much improvement in bias and so generally isn't used.

# Nonlinear Regression

Levenberg-Marquardt optimizes the maximum likelihood estimation by searching for the sum of squares minima of a multivariate distrubution through a combination of gradient descent and Gauss-Newton (parabolic) optimization.  Iterates, at each step taking the partial derivatives of all parameters and miniming the sum of squares between the function and the partial derivatives times a small delta in the params, generally taking smaller steps with increasing iterations.

```python
from scipy import optimize

x = np.linspace(-3,3,100) # 100 values between -3 and 3

# We will use first 6 terms of a taylor series expansion to approximate sin(x)
#  where sin(x) ~= sum(a_n x^n)
#
def taylor_err(a, x, y):
    p = np.arange(len(a))[:, np.newaxis] # column vector
    return y - np.dot(a, x ** p)

# Actual values of the function at x
y = np.sin(x)

a_start = np.zeros(6)               # initial guess at the parameters
a_best, flag = optimize.leastsq(taylor_err, a_start, args=(x, y))  # fit using LM

y_fit = np.dot(a_best, x ** np.arange(len(a_best))[:, np.newaxis]) # y values of our model at x

a_best # just show our 5th degree polynomial fit
```

```python
# plot the sin function and our LM fit approximation
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)

ax1.set_xlabel(r'$x$')
ax1.set_ylabel(r'$y$')

pSin = ax1.plot(x, y, linestyle='-', color='#cccccc',linewidth=3)    # Actual sin function
pFit = ax1.plot(x, y_fit, linestyle='-.', color='#cc0000')
```

<i>Warning</i>: fit often depends on initial parameters being close to avoid local minima.
