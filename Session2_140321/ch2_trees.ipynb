## Quad and Oct trees

First, we need to set up a structure to store the tree.  To do this we define a class of objects called 'QuadTree's.  To do this I have stoeln the code from AstroMl used to create figure 2.3.  I have gone through and added some comments in here.  I think it is very useful to go through and try to understand everything that is going on here.

```python
class QuadTree:
    """Simple Quad-tree class"""

    # class initialization function
    def __init__(self, data, mins, maxs, depth=3):
        self.data = np.asarray(data) #this object will have a property called data, which is data array passed in when making the tree

        # data should be two-dimensional
        assert self.data.shape[1] == 2 #remember that quad trees work on 2-d data

        #define the range the data falls within
        if mins is None:
            mins = data.min(0)
        if maxs is None:
            maxs = data.max(0)

        #give the new tree object some properties...
        self.mins = np.asarray(mins)
        self.maxs = np.asarray(maxs)
        self.sizes = self.maxs - self.mins

        self.children = [] #an empty list

        mids = 0.5 * (self.mins + self.maxs) #dividing the range in two
        xmin, ymin = self.mins #a fancy way of writing xmin=self.mins[0] and ymin=self.mins[1]
        xmax, ymax = self.maxs
        xmid, ymid = mids

        if depth > 0:
            # split the data into four quadrants
            data_q1 = data[(data[:, 0] < mids[0]) & (data[:, 1] < mids[1])]
            data_q2 = data[(data[:, 0] < mids[0]) & (data[:, 1] >= mids[1])]
            data_q3 = data[(data[:, 0] >= mids[0]) & (data[:, 1] < mids[1])]
            data_q4 = data[(data[:, 0] >= mids[0]) & (data[:, 1] >= mids[1])]

            # recursively build a quad tree on each quadrant which has data
            if data_q1.shape[0] > 0: self.children.append(QuadTree(data_q1, [xmin, ymin], [xmid, ymid], depth - 1))
            if data_q2.shape[0] > 0: self.children.append(QuadTree(data_q2, [xmin, ymid], [xmid, ymax], depth - 1))
            if data_q3.shape[0] > 0: self.children.append(QuadTree(data_q3, [xmid, ymin], [xmax, ymid], depth - 1))
            if data_q4.shape[0] > 0: self.children.append(QuadTree(data_q4, [xmid, ymid], [xmax, ymax], depth - 1))
    
    #they define this useful little function to make plotting the quad tree easier
    def draw_rectangle(self, ax, depth):
        """Recursively plot a visualization of the quad tree region"""
        if depth is None or depth == 0:
            rect = plt.Rectangle(self.mins, *self.sizes, zorder=2, ec='#000000', fc='none') #draw a black rectangle
            ax.add_patch(rect)
        if depth is None or depth > 0:
            for child in self.children:
                child.draw_rectangle(ax, depth - 1)
                
                
```

Now that we have defined a quad tree structure, we can use it!  Let's create some data... how about 30 (x,y) pairs.

```python
np.random.seed(0) #this initializes the random number generator.

X = np.random.random((30, 2)) #30 rows of 2 random numbers between 0 and 1
print X
plt.plot(X[:,0],X[:,1],'o') #lets plot it to make sure it is what we think it is... btw X[:,0] accesses the first column
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

now lets add some structure to these points, as they are pretty boring as it stands.

```python
X = X * 2 - 1
X[:, 1] *= 0.1
X[:, 1] += X[:, 0] ** 2

plt.plot(X[:,0],X[:,1],'o')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

now we can use the tree structure we defined to build one for this data set:

```python
mins = (-1.1, -0.1)
maxs = (1.1, 1.1)
QT = QuadTree(X, mins, maxs, depth=3)
```

now we can look at the tree visually by plotting the different levels of the tree.

```python
fig = plt.figure(figsize=(10, 10)) #create a 5 inch by 5 inch space
fig.subplots_adjust(wspace=0.1, hspace=0.15, left=0.1, right=0.9, bottom=0.05, top=0.9) #define the location of the plotting box

def draw_grid(ax, xlim, ylim, Nx, Ny, **kwargs): #define a function to draw a grid
    """ draw a background grid for the quad tree"""
    for x in np.linspace(xlim[0], xlim[1], Nx):
        ax.plot([x, x], ylim, **kwargs)
    for y in np.linspace(ylim[0], ylim[1], Ny):
        ax.plot(xlim, [y, y], **kwargs)

for level in range(1, 5): #run through each level: 1,2,3,4
    ax = fig.add_subplot(2, 2, level, xticks=[], yticks=[]) # add a subplot at location "level"
    ax.scatter(X[:, 0], X[:, 1])
    QT.draw_rectangle(ax, depth=level - 1) #draw a rectangle 

    Nlines = 1 + 2 ** (level - 1)
    draw_grid(ax, (mins[0], maxs[0]), (mins[1], maxs[1]), Nlines, Nlines, linewidth=1, color='#CCCCCC', zorder=0) #draw a grey grid

    ax.set_xlim(-1.2, 1.2)
    ax.set_ylim(-0.15, 1.15)
    ax.set_title('level %i' % level)

# suptitle() adds a title to the entire figure
fig.suptitle('Quad-tree Example')
plt.show()
```
